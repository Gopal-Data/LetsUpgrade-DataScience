```python
import pandas as pd
from dateutil.parser import parse
date_series = pd.Series(['01 Jan 2015', '10-02-2016', '20180307', '2014/05/06', '2016-04-12', '2019-04-06T11:20'])
date_series = date_series.map(lambda x: parse(x))
print('Day of month:',date_series.dt.day.tolist())
print("Week number:",date_series.dt.weekofyear.tolist())
print("Day of year:",date_series.dt.dayofyear.tolist())
print("Day of week:",date_series.dt.day_name().tolist())
```

```python
x = pd.Series([1, 2, 3, 4, 5]) 
y = pd.Series([6, 7, 8, 9, 10]) 
# here we are computing every thing 
# step by step 
p1 = np.sum([(a * a) for a in x]) 
p2 = np.sum([(b * b) for b in y]) 
  
# using zip() function to create an iterator which aggregates elementsfrom two or more iterables 
p3 = -1 * np.sum([(2 * a*b) for (a, b) in zip(x, y)]) 
dist = np.sqrt(np.sum(p1 + p2 + p3)) 
print("Euclidean distance between two series is:", dist) 
```

```python
equake_sort=earthquake.sort_values(by=['mag'],ascending=False).head(20)
equake_sort
```

```python
earthquake['place'].value_counts().head(10)
```

```python
dataa = earthquake["place"].str.split(",",n=1, expand = True) #create a New Column state and location with place column
```

```python
earthquake = pd.concat([earthquake,dataa],axis=1)
earthquake.rename(columns={0:'Location',1:'State'},inplace=True)
earthquake.drop_duplicates(keep='first')
earthquake.head(5)
```

```python
Stateonly=earthquake['State']
Stateonly
```

```python
top10state=Stateonly.groupby(by=Stateonly).agg(['nunique','size'])
top10state = top10state.sort_values('size', ascending=[False])
top10state.head(10)
```

```python
from sklearn import datasets
iris = datasets.load_iris()
irisf = pd.DataFrame(iris.data)
irisf.columns = ['Sepal_length','sepal_width','petal_length','Petal_width']
irist= pd.DataFrame(iris.target)
irist.columns = ['Target']
iris=pd.concat([irisf, irist], axis = 1)
iris.head()
```

```python
iris.describe()
```

```python
import matplotlib.pyplot as plt
plt.hist('Target',data=iris);
```

```python
labels = ['setosa', 'versicolor', 'virginica']
X_pie=iris['Target'].value_counts()
plt.pie(X_pie,labels=labels,autopct='%.2f%%'); 
```

```python
import seaborn as sns
sns.catplot(x='Target', y='Sepal_length', kind='box',data=iris)
sns.stripplot(x='Target',y='Sepal_length',data=iris);
```

```python
cor_relation=iris.corr()
sns.heatmap(cor_relation, annot=True);
```
